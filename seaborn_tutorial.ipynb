```python
#faceGrid
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

df = sns.load_dataset('tips')
print(df.to_string())

g = sns.FacetGrid(df, col='time')
g.map(sns.distplot, 'total_bill')
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('tips')
g = sb.FacetGrid(df, col="sex", hue="smoker")
g.map(plt.scatter, "total_bill", "tip")
plt.show()
```

```python
#stastical estmation
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('tips')
sb.barplot(x="sex",y="tip",hue="time",data=df)
plt.show()
```

```python
#linear relationship
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('tips')
sb.regplot(x="total_bill",y="tip",data=df)
plt.show()
```

```python
sb.pairplot(df)
plt.show()
```

```python
selected_vars = ['total_bill', 'tip']
sb.pairplot(df[selected_vars])
plt.show()
```

```python
sb.pairplot(df,hue='size',palette='husl')
plt.show()
```

```python
#diagonal kinf plots
sb.pairplot(df,diag_kind='kde')
plt.show()
```

```python
#adjusting the plot kind
sb.pairplot(df,kind='reg')
plt.show()
```
